```python
import numpy as np
import os
import re
```

```python
# Tomek's simslist
grid_runs_dir = "/Users/arne/git/cosmogrid_desy3/data/CosmoGrid_grid_simslist.npy"
grid_runs = np.load(grid_runs_dir)["path_sim"]

print(grid_runs[:5])
```

```python
cosmos = [re.search("cosmo_[0-9]{6}", run).group(0) for run in grid_runs]
assert len(cosmos) == 7*2500

unique_cosmos = np.unique(cosmos)
assert len(unique_cosmos) == 2500
print(unique_cosmos[:10])
```

```python
grid_perms = []
for cosmo in unique_cosmos:
    for i in range(4):
        grid_perms.append(os.path.join(cosmo, f"perm_{i:04}"))

print(grid_perms[:10])
assert len(grid_perms) == 4*2500
```

```python
grid_perms_dir = "/Users/arne/git/cosmogrid_desy3/data/CosmoGrid_perms_list"
np.save(grid_perms_dir, grid_perms)
```
