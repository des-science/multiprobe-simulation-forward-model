# pipeline

```python
grid_pipe = GridPipeline(
    conf=conf,
    params=params,
    with_lensing=True,
    with_clustering=True,
    with_padding=True,
    apply_norm=False
)
```

```python
data_vec_pix = grid_pipe.data_vec_pix
n_side = conf["analysis"]["n_side"]
n_pix = conf["analysis"]["n_pix"]
n_z = grid_pipe.n_z_metacal + grid_pipe.n_z_maglim
n_z_metacal = grid_pipe.n_z_metacal
n_noise = conf["analysis"]["grid"]["n_noise_per_example"]

n_patches = conf["analysis"]["n_patches"]
n_perms_per_cosmo = conf["analysis"]["grid"]["n_perms_per_cosmo"]
n_noise_per_example = conf["analysis"]["grid"]["n_noise_per_example"]
n_examples_per_cosmo = n_patches * n_perms_per_cosmo * n_noise_per_example

n_dset_items = 1000
```

### default, evaluation mode

```python
eval_dset = grid_pipe.get_dset(
    tfr_pattern=tfr_pattern,
    n_noise=n_noise,
    local_batch_size=1,
    n_readers=1,
    n_prefetch=0,
)

counter = 0
for dv, cosmo, index in eval_dset.take(n_dset_items):
    if counter % n_examples_per_cosmo == 0:
        print(np.squeeze(index[0].numpy()))
    counter += 1
```
