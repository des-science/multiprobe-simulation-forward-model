```python
import numpy as np
import healpy as hp
import matplotlib.pyplot as plt

from msfm.utils import scales
```

```python
n_side = 512
l = np.arange(3*n_side-1)
```

```python
l_min = 30
l_max = 800

low_pass_fac = scales.gaussian_low_pass_factor_alm(l, l_max=l_max)
high_pass_fac = scales.gaussian_high_pass_factor_alm(l, l_min=l_min)

fig, ax = plt.subplots()

ax.plot(l, low_pass_fac, label="low pass, removes small scales")
ax.plot(l, high_pass_fac, label="high pass, removes big scales")

ax.legend()
ax.set(xlabel=r"$\ell$", ylabel=r"alm factor")
```
