```python
import numpy as np
import healpy as hp
from scipy.integrate import quad
import matplotlib.pyplot as plt
```

```python
l_max = 700
fwhm = np.pi/l_max
sigma = fwhm/2.355

standard = lambda x: 1/(sigma * np.sqrt(2*np.pi)) * np.exp(-0.5 * (x/sigma)**2)
```

```python
x = np.linspace(-3*sigma, 3*sigma, 1000)

fig, ax = plt.subplots()
ax.plot(x, standard(x))
ax.axvline(fwhm/2, color='k', linestyle='--')
ax.axvline(- fwhm/2, color='k', linestyle='--')
```
