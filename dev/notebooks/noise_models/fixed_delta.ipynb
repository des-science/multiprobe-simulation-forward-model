### mean over N realizations

```python
noise_maps = [rot_noise, emp_noise]
noise_types = ["rotate in place", "empirical distribution"]
component = "g1"
N = len(rot_noise[component])

all_noise_maps = np.stack([noise_map[component] for noise_map in noise_maps])
bins = np.linspace(all_noise_maps.min(), all_noise_maps.max(), 50)
```

```python
fig_cls, ax_cls = plt.subplots(figsize=(12,8))

fig_hist, ax_hist = plt.subplots(figsize=(12,8))

for noise_map, noise_type in zip(noise_maps, noise_types):
    noise_map = noise_map[component]
    
    cls = []
    hists = []
    for i in range(N):
        current_noise_map = noise_map[i]
        current_noise_patch = current_noise_map[current_noise_map!=0]

        cls.append(hp.anafast(current_noise_map, use_pixel_weights=True))

        hist, bin_edges = np.histogram(current_noise_patch, bins=bins)
        hists.append(hist)

    # cls
    cls_mean = np.mean(cls, axis=0)
    cls_std = np.std(cls, axis=0)/np.sqrt(N)

    l = np.arange(1, l_max)
    ax_cls.loglog(l, cls_mean[l], label=noise_type)
    ax_cls.fill_between(l, cls_mean[l]-cls_std[l], cls_mean[l]+cls_std[l], alpha=0.2)

    # histogram
    hist_mean = np.mean(hists, axis=0)
    hist_std = np.std(hists, axis=0)

    ax_hist.bar(bins[:-1], hist_mean, width=np.diff(bins), edgecolor="k", align="edge", label=noise_type, alpha=0.3, log=True, yerr=hist_std)


ax_cls.legend()
ax_cls.set(xlabel=r"$\ell$", ylabel="Cls")
ax_cls.grid(True)

ax_hist.legend()
ax_cls.set(xlabel="ellipticity", ylabel="number of pixels")
ax_cls.grid(True)
```
