```python
import numpy as np
import healpy as hp
```

```python
n_side = 512
n_pix = hp.nside2npix(n_side)

in_map = np.ones(n_pix)
```

```python
%%timeit
alm = hp.map2alm(in_map)
```

```python
from time import time

n_iterations = 100
t0 = time()
for i in range(n_iterations):
    alm = hp.map2alm(in_map)
    print(t0 - time())
    t0 = time()
    
ttot = (time() - t0)/n_iterations
print(ttot)
```
