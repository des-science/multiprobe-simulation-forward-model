```python
import numpy as np
import healpy as hp
```

```python
n_side = 512
A_pix = hp.nside2pixarea(n_side, degrees=True)
```

# Metacal

```python
# from Table 1 in https://arxiv.org/pdf/2105.13543.pdf, in units of galaxies/deg^2
n_gal = np.array([1.476, 1.479, 1.484, 1.461])*60**2
print(n_gal)

# units of galaxies/pixel
n_gal *= A_pix
print(n_gal)
```

# Maglim

```python
# from Table 1 in https://arxiv.org/pdf/2105.13540.pdf, in units of galaxies/deg^2
n_gal = np.array([0.15, 0.107, 0.109, 0.146, 0.106, 0.1])*60**2
print(n_gal)

# units of galaxies/pixel
n_gal *= A_pix
print(n_gal)
```
